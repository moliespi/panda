```python
#Diagrama de lineas
fig, ax = plt.subplots()
ax.plot([1, 2, 3, 4], [1, 2, 0, 0.5])
plt.show()

fig, ax = plt.subplots()
ax.plot([1, 4, 1, 4], [1, 1, 0, 0])
plt.show()
```

```python
peso1=input("Ingresa tu peso en KG: ")
peso2=input("Ingresa tu peso en KG: ")
peso3=input("Ingresa tu peso en KG: ")
peso4=input("Ingresa tu peso en KG: ")
peso5=input("Ingresa tu peso en KG: ")
pesos=[int(peso1),int(peso2),int(peso3),int(peso4),int(peso5)]

fig, ax = plt.subplots()
ax.pie(pesos, labels=pesos)
plt.show()

#promedio de edades
promedio=np.mean(pesos)
#Desviacion estandar de edades
desviacion=np.std(pesos)

#mostrar en un grafico el promedio y la desviacion
fig, ax = plt.subplots()
ax.pie([promedio,desviacion ], labels=["Promedio", "Desviacion"], autopct='%1.1f%%')
plt.show()

```

```python
#Porcentaje de mujeres y varones en un grafico
mujeres=input("Ingresa el numero de mujeres:")
varones=input("Ingresa el numero de varones: ")

total = int(mujeres) + int(varones)
#procentaje de la cantidad de mujeres y varones
procentaje_varones=(int(varones)/total)*100
porcentaje_mujeres=(int(mujeres)/total)*100

fig, ax =plt.subplots()
ax.pie([procentaje_varones,porcentaje_mujeres ], labels=["Varones", "Mujeres"], autopct='%1.1f%%')
ax.set_title('Porcentaje de hombres y mujeres', loc = "center", fontdict = {'fontsize':14, 'fontweight':'bold', 'color':'tab:blue'})
plt.show()
```
